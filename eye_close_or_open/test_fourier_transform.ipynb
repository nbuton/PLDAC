```python
import numpy as np
from numpy.fft import fft,fftfreq
import matplotlib.pyplot as plt
```

```python
n = 1000
X = np.linspace(0,4,n)
#Signal de periode 1 et d'amplitude 2
y1 = 2*np.sin(1*X*2*np.pi)
#Signal de periode 3 et d'amplitude 4
y2 = 4*np.sin(3*X*2*np.pi)
plt.plot(X,y1)
```

```python
plt.plot(X,y2)
```

```python
Y = y1#Y = y1+y2
plt.plot(X,Y)
```

```python
def spectre(signal):
    fft = np.abs(np.fft.fft(signal))
    freqs = np.fft.fftfreq(len(signal))
    moitier = int(len(fft)/2)
    fft = fft[:moitier]
    freqs = freqs[:moitier]
    print(len(fft))
    print(len(freqs))
    plt.plot(freqs,fft)
```

```python
spectre(Y)
```

```python
a=spectre(Y)
whereMax = np.r_[True, a[1:] < a[:-1]] & np.r_[a[:-1] < a[1:], True]
for k in range(len(whereMax)):
    if(whereMax[k]):
        print(k)
        print(a[k])
```

```python
import numpy as np
from numpy.fft import fft,fftfreq
import matplotlib.pyplot as plt
```

```python
n = 100
X = np.linspace(0,5,n)
y1 = 2*np.sin(3*X)
y2 = 4*np.sin(5*X)
Y = y1+y2
plt.plot(X,Y)
```

```python
def spectre(signal):
    freqs = fftfreq(len(signal))
    mask = freqs >0
    fft_val = fft(signal)
    fft_theo = 2.0*np.abs(fft_val/n)
    return fft_theo[mask]
```

```python
plt.plot(spectre(Y))
```
